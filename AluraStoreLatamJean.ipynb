### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Facturación total por tienda
facturacion_tienda1 = tienda['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

print(f'Facturación Tienda 1: ${facturacion_tienda1:,.2f}')
print(f'Facturación Tienda 2: ${facturacion_tienda2:,.2f}')
print(f'Facturación Tienda 3: ${facturacion_tienda3:,.2f}')
print(f'Facturación Tienda 4: ${facturacion_tienda4:,.2f}')
```

```python
# Gráfico de facturación
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturaciones = [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4]

plt.figure(figsize=(8, 5))
plt.bar(tiendas, facturaciones, color=['#4C72B0', '#DD8452', '#55A868', '#C44E52'])
plt.title('Facturación Total por Tienda')
plt.ylabel('Facturación (COP)')
plt.xlabel('Tienda')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Ventas por categoría — cantidad de productos vendidos por tienda
cat_tienda1 = tienda['Categoría del Producto'].value_counts()
cat_tienda2 = tienda2['Categoría del Producto'].value_counts()
cat_tienda3 = tienda3['Categoría del Producto'].value_counts()
cat_tienda4 = tienda4['Categoría del Producto'].value_counts()

print('Ventas por categoría — Tienda 1:')
print(cat_tienda1)
print('\nVentas por categoría — Tienda 2:')
print(cat_tienda2)
print('\nVentas por categoría — Tienda 3:')
print(cat_tienda3)
print('\nVentas por categoría — Tienda 4:')
print(cat_tienda4)
```

```python
# Gráfico de ventas por categoría (Tienda 1 como referencia)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Ventas por Categoría del Producto', fontsize=14)

for ax, cat, titulo in zip(
    axes.flatten(),
    [cat_tienda1, cat_tienda2, cat_tienda3, cat_tienda4],
    ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
):
    ax.bar(cat.index, cat.values)
    ax.set_title(titulo)
    ax.set_ylabel('Cantidad de ventas')
    ax.set_xticklabels(cat.index, rotation=45, ha='right', fontsize=8)

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio por tienda
calif_tienda1 = tienda['Calificación'].mean()
calif_tienda2 = tienda2['Calificación'].mean()
calif_tienda3 = tienda3['Calificación'].mean()
calif_tienda4 = tienda4['Calificación'].mean()

print(f'Calificación promedio Tienda 1: {calif_tienda1:.2f}')
print(f'Calificación promedio Tienda 2: {calif_tienda2:.2f}')
print(f'Calificación promedio Tienda 3: {calif_tienda3:.2f}')
print(f'Calificación promedio Tienda 4: {calif_tienda4:.2f}')
```

```python
# Gráfico de calificación promedio
calificaciones = [calif_tienda1, calif_tienda2, calif_tienda3, calif_tienda4]

plt.figure(figsize=(8, 5))
plt.bar(tiendas, calificaciones, color=['#4C72B0', '#DD8452', '#55A868', '#C44E52'])
plt.title('Calificación Promedio por Tienda')
plt.ylabel('Calificación (1 a 5)')
plt.xlabel('Tienda')
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Productos más y menos vendidos por tienda
for nombre, df in zip(tiendas, [tienda, tienda2, tienda3, tienda4]):
    conteo = df['Producto'].value_counts()
    print(f'\n=== {nombre} ===')
    print(f'  Más vendido:  {conteo.idxmax()} ({conteo.max()} ventas)')
    print(f'  Menos vendido: {conteo.idxmin()} ({conteo.min()} ventas)')
```

```python
# Gráfico de top 5 más y menos vendidos por tienda
fig, axes = plt.subplots(4, 2, figsize=(14, 18))
fig.suptitle('Productos Más y Menos Vendidos por Tienda', fontsize=14)

for i, (nombre, df) in enumerate(zip(tiendas, [tienda, tienda2, tienda3, tienda4])):
    conteo = df['Producto'].value_counts()

    # Más vendidos
    top5 = conteo.head(5)
    axes[i][0].bar(top5.index, top5.values, color='steelblue')
    axes[i][0].set_title(f'{nombre} — Más vendidos')
    axes[i][0].set_xticklabels(top5.index, rotation=45, ha='right', fontsize=8)
    axes[i][0].set_ylabel('Ventas')

    # Menos vendidos
    bot5 = conteo.tail(5)
    axes[i][1].bar(bot5.index, bot5.values, color='salmon')
    axes[i][1].set_title(f'{nombre} — Menos vendidos')
    axes[i][1].set_xticklabels(bot5.index, rotation=45, ha='right', fontsize=8)
    axes[i][1].set_ylabel('Ventas')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Costo de envío promedio por tienda
envio_tienda1 = tienda['Costo de envío'].mean()
envio_tienda2 = tienda2['Costo de envío'].mean()
envio_tienda3 = tienda3['Costo de envío'].mean()
envio_tienda4 = tienda4['Costo de envío'].mean()

print(f'Costo de envío promedio Tienda 1: ${envio_tienda1:,.2f}')
print(f'Costo de envío promedio Tienda 2: ${envio_tienda2:,.2f}')
print(f'Costo de envío promedio Tienda 3: ${envio_tienda3:,.2f}')
print(f'Costo de envío promedio Tienda 4: ${envio_tienda4:,.2f}')
```

```python
# Gráfico de costo de envío promedio
envios = [envio_tienda1, envio_tienda2, envio_tienda3, envio_tienda4]

plt.figure(figsize=(8, 5))
plt.bar(tiendas, envios, color=['#4C72B0', '#DD8452', '#55A868', '#C44E52'])
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo de envío (COP)')
plt.xlabel('Tienda')
plt.tight_layout()
plt.show()
```
